```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
pd.set_option("display.max_columns", None)
#df = pd.read_csv("House_Rent_Dataset.csv")
df = pd.read_csv("/content/drive/MyDrive/#2-M.Sc Computer Science [SXC]/1st Year/Books and notes/Sem 4/Machine Learning/SM+SS/13-02-23/House_Rent_Dataset.csv")
df.head()
```

```python
df.info()
```

```python
target = list()
for bhk in df['BHK']:
  if bhk < 2:
    target.append(0)
  elif bhk >= 2 and bhk < 4:
    target.append(1)
  else:
    target.append(2)

target[:5]
```

```python
df["target"] = target
df.head()
```

```python
area_type_dict = {
    'Super Area': 0, 'Carpet Area': 1, 'Built Area': 2
}

fur_status_dict = {
    'Semi-Furnished': 0, 'Unfurnished': 1, 'Furnished': 2
}

city_dict = {
 'Kolkata': 0,
 'Bangalore': 1,
  'Mumbai': 2,
 'Chennai': 3,
 'Hyderabad': 4,
 'Delhi': 5,
}

tenant_dict = {
    'Bachelors/Family': 0,
    'Bachelors': 1,
    'Family': 2
}

contact_dict = {
    'Contact Owner': 0,
    'Contact Agent': 1,
    'Contact Builder': 2
}
```

```python
df['Point of Contact'].value_counts()
```

```python
df['Point of Contact'] = df['Point of Contact'].apply(lambda x: contact_dict[x])
```

```python
df['Tenant Preferred'] = df['Tenant Preferred'].apply(lambda x: tenant_dict[x])
```

```python
df['Area Type'] = df['Area Type'].apply(lambda x: area_type_dict[x])
df['Furnishing Status'] = df['Furnishing Status'].apply(lambda x: fur_status_dict[x])
df['City'] = df['City'].apply(lambda x: city_dict[x])
```

```python
df.head()
```

```python
area_dict = df['Area Locality'].value_counts().to_dict()
```

```python
for i, l in enumerate(area_dict):
  area_dict[l] = i
```

```python
df['Area Locality'] = df['Area Locality'].apply(lambda x: area_dict[x])
```

```python
fl_dict = df['Floor'].value_counts().to_dict()
for i, l in enumerate(fl_dict):
  fl_dict[l] = i
```

```python
df['Floor'] = df['Floor'].apply(lambda x: fl_dict[x])
```

```python
df.corr()
```

```python
df = df.drop(columns=["BHK", "Posted On"], axis="columns")
```

```python
df.head()
```

```python
X = df.iloc[:, :-1]
Y = df.iloc[:, -1]
```

```python
X_train, X_test, Ytrain, Y_test = train_test_split(X, Y, test_size=0.2, random_state=43)
```

```python
svm = SVC()
svm.fit(X_train, Ytrain)
```

```python
pred = svm.predict(X_test)
```

```python
svm.score(X_test, Y_test)
```

```python
svm = SVC(kernel="poly")
svm.fit(X_train, Ytrain)
```

```python
svm.score(X_test, Y_test)
```

```python
dTree = DecisionTreeClassifier(criterion = "entropy")
dTree.fit(X_train, Ytrain)
```

```python
dTree.score(X_test, Y_test)
```

```python
dTree = DecisionTreeClassifier(criterion = "gini")
dTree.fit(X_train, Ytrain)
```

```python
dTree.score(X_test, Y_test)
```
