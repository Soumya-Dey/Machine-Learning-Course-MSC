```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
# data = pd.read_csv("/content/drive/MyDrive/#2-M.Sc Computer Science [SXC]/1st Year/Books and notes/Sem 4/Machine Learning/SM+SS/27-02-23/diabetes.csv")
data = pd.read_csv("diabetes.csv")
data.head()
```

```python
sns.heatmap(data.corr())
plt.title('Correlation On iris Classes')
```

```python
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,8]
```

```python
x_train,x_test, y_train, y_test=train_test_split(x,y,test_size=0.1, random_state=43)
```

```python
model = SVC()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, pred))
```

```python
"{0:.2f}%".format(accuracy_score(y_test, pred) * 100)
```

```python
cm = confusion_matrix(y_test, pred)
cm
```

```python
sns.heatmap(cm, annot=True, cmap="PuBu")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.show()
```
