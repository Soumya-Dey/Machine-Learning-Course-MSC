- #### Multiple Regression

```python
df_n = df[['Size', 'Furnishing Status', 'Rent']]
df_n.head()
```

```python
df_n = pd.get_dummies(df_n)
```

```python
df_n.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df_n[['Size', 'Furnishing Status_Furnished', 'Furnishing Status_Semi-Furnished', 'Furnishing Status_Unfurnished']]
Y = df_n[['Rent']]
```

```python
regr = LinearRegression()
regr.fit(X, Y)
```

```python
regr.predict([[1100, 0, 1, 0]])
```

- #### Polynomial Regression

```python
myModel = np.poly1d(np.polyfit(df_n['Size'], df['Rent'], 3))
myLine = np.linspace(df_n['Size'].min(), df_n['Size'].max(), 500)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df_n['Size'], df['Rent'])
plt.plot(myLine, myModel(myLine))
plt.show()
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(df['Rent'], myModel(df_n['Size']))
```

- #### Linear Regression

```python
from scipy import stats
linregr = stats.linregress(df_n['Size'], df['Rent'])
```

```python
def myFunc(x):
    return (linregr.slope * x) + linregr.intercept
```

```python
myModel = [myFunc(i) for i in df_n['Size']]
```

```python
plt.scatter(df_n['Size'], df['Rent'])
plt.plot(df_n['Size'], myModel)
plt.show()
```
