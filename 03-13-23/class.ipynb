[K-Means](https://towardsdatascience.com/k-means-clustering-algorithm-applications-evaluation-methods-and-drawbacks-aa03e644b48a?gi=0bccff4f37a1)
- Applied on numerical data which can be divided into clusters
- We minimize the variance in each cluster
- Hierarchical clustering??
- `(point_1 - mean)^2 + (point_2 - mean)^2 + (point_3 - mean)^2`
- Working:
  - First, each data point is randomly assigned to one of the K clusters. Then, we compute the centroid (functionally the center) of each cluster and reassign each data data point to the cluster with the closest centroid. We repeat this.
  - K-means clustering requires us to select K, the  no. of clusters we want to group the data into. The elbow method lets us graph the inertia and visualize the point at which it starts decreasing linearly. This point is referred to as the 'Elbow' and is a good estimate for beat value of K based on it.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x = [1, 2, 5, 7, 8, 9, 6, 65, 85, 70]
y = [5, 6, 7, 8, 9, 78, 65, 7, 23, 2]
data = list(zip(x, y))

inertia = []
for i in range (1, 11):
  kmeans = KMeans(n_clusters = i)
  kmeans.fit(data)
  inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker = 'o')
plt.xlabel("No. of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
X= -2 * np.random.rand(100,2)
X1 = 1 + 2 * np.random.rand(50,2)
X[50:100, :] = X1
plt.scatter(X[ : , 0], X[ :, 1], s = 50, c = 'b')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(X[ : , 0], X[ : , 1], s =50, c='b')
plt.scatter(1.91041875,  1.77019919, s=200, c='g', marker='s')
plt.scatter(-1.02421204, -0.95085292, s=200, c='r', marker='s')
plt.show()
```

```python
kmeans.labels_
```

```python
sample_test=np.array([-3.0,-3.0])
second_test=sample_test.reshape(1, -1)
kmeans.predict(second_test)
```
