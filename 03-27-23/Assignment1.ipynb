```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
df = pd.read_csv("/content/drive/MyDrive/#2-M.Sc Computer Science [SXC]/1st Year/Books and notes/Sem 4/Machine Learning/SM+SS/27-03-23/employees.csv")
df.head()
```

```python
df = df.drop(columns=["First Name", "Last Login Time"], axis=1)
```

```python
df.head()
```

```python
df['Gender'].unique()
```

```python
df['Gender'] = df['Gender'].fillna("Other")
```

```python
df['Gender'].unique()
```

```python
gender_dict = {
    "Male": 0,
    "Female": 1,
    "Other": 2
}
df['Gender'] = df['Gender'].apply(lambda el: gender_dict[el])
```

```python
df['Gender'].unique()
```

```python
df['Senior Management'] = df['Senior Management'].apply(lambda el: 1 if el else 0)
```

```python
df.head()
```

```python
df['Team'] = df['Team'].fillna("Other")
```

```python
team_dict = {}
for i, el in enumerate(df['Team'].unique()):
  team_dict[el] = i
```

```python
team_dict['Other'] = 0
team_dict['Marketing'] = 1
```

```python
team_dict
```

```python
df['Team'] = df['Team'].apply(lambda el: team_dict[el])
```

```python
df.head()
```

```python
df['Year'] = df['Start Date'].apply(lambda el: el.split("/")[-1])
```

```python
df = df.drop(columns=['Start Date'], axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Year'] = df['Year'].astype("int64")
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['Salary'], axis=1)
Y = df['Salary']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=43)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
new_data = [[1, 11.76, 0, 2, 1998], [0, 5.88, 1, 1, 1992], [1, 2.00, 0, 0, 2023]]
pred = model.predict(new_data)
pred
```

```python
model2 = SVR(kernel = 'linear')
```

```python
model2.fit(x_train, y_train)
```

```python
model2.score(x_test, y_test)
```

```python
new_data = [[1, 11.76, 0, 2, 1998], [0, 5.88, 1, 1, 1992], [1, 2.00, 0, 0, 2023]]
pred = model2.predict(new_data)
pred
```
